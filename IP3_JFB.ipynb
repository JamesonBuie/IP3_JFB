```python
X = np.linspace(0, 100, 101)
```

```python
X_2=X**2
```

```python
print(X_2)
```

```python
print(X)
```

```python
fig,ax1=plt.subplots()
fig=plt.figure(figsize=(12,6))
# Set x label to the date variable
ax1.set_xlabel('X vs X squared')
# disabling ticks and labels from bottom
ax1.tick_params(axis='x',bottom=False, labelbottom=False)

#setting y label
ax1.set_ylabel('X',color='firebrick',size='x-large')
ax1.tick_params(axis='y',labelcolor='firebrick',labelsize='large')

ax1.plot(X,color='firebrick')


ax2=ax1.twinx() #creates a twin of ax1
ax2.set_ylabel('X squared',color='cornflowerblue',size='x-large')
ax2.tick_params(axis='y',labelcolor='cornflowerblue',labelsize='large')

ax2.plot(X_2,color='cornflowerblue')

plt.show()
```

```python

fig, ax = plt.subplots()
ax.plot(X, X_2, label='X^2')


ax.set_xlabel('X')
ax.set_ylabel('X^2')
ax.set_title('X versus X^2')

plt.show()
```

# 16

```python
heat_map=carsales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
cols=['Total Sales','Price','Odometer (KM)']
sns.heatmap(heat_map,xticklabels=cols,yticklabels=cols,annot=True)
plt.show()
```

The correlation between total sales and price is very weak positive relationship with a value of .16. This means that price and total sales do not have a linear relationship that would be explained by a linear model very well. Next, total sales and the odometer variables have a correlation value of -0.32. This shows that they have a fairly weak, negative linear relationship. They do not have a very strong impact on each other. Lastly, price and the odometer variable have the highest strongest correlation of -0.63. This means that they have a fairly strong, negative linear relationship. These two variables have the most significant impact on each other. As the value of one variable goes up, the value of the other variable goes down.
